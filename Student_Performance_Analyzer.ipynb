```python
import pandas as pd

data={
    "Name":["Laila","ALi","Rubii","Armaan"],
    "GPA":[3.44,3.67,2.0,1.90],
    "Department":["AI","CS","AI","DS"]
}

df = pd.DataFrame(data)

print(df)

def get_status(gpa):
    if gpa>=3:
        return "Dean's list"
    elif gpa>=2:
        return "PASS"
    else:
        return"FAIL"
df["Status"]=df["GPA"].apply(get_status)
print(df)

average_dep=df.groupby("Department")["GPA"].mean()
print(average_dep)

```

```python
print("="*40)
print(" STUDENT PERFORMANCE REPORT")
print("="*40)
print(f"Total Students:{len(df)}")
print(f"Highest GPA:{df["GPA"].max()}")
print(f"Average GPA:{df["GPA"].mean()}")
print(f"Best Department:{average_dep.idxmax()}")
print(f"Highest Department:{average_dep.max()}")
print("="* 40)
```
