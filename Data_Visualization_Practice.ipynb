```python
import matplotlib.pyplot as plt

names = ["Laila", "Ali", "Rubi", "Armaan"]
gpas  = [3.44, 3.67, 2.0, 1.90]

# Chart 1 — Bar chart
plt.bar(names, gpas, color=["red", "green", "black", "yellow"])
plt.title("Student GPA Comparison")
plt.xlabel("Student Names")
plt.ylabel("GPA")
plt.ylim(0, 4.5)
for i, gpa in enumerate(gpas):
    plt.text(i, gpa + 0.05, str(gpa), ha="center")
plt.show()

# Chart 2 — Line chart
plt.plot(names, gpas, color="blue", marker="^")
plt.title("GPA Trend")
plt.xlabel("Student Names")
plt.ylabel("GPA")
plt.ylim(0, 4.5)

for i,gpa in enumerate(gpas):
    plt.text(names[i],gpa+0.05,str(gpa),ha="center")
plt.show()

# Chart 3 — Pie chart
statuses = ["Dean's List", "PASS", "FAIL"]
counts =   [2, 1, 1]
plt.pie(counts, labels=statuses, autopct="%1.1f%%")
plt.title("Status Breakdown")
plt.show()
```
